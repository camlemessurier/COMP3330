# Introduction to SVM Parameters

This example illustrates the effect of the parameters gamma and C of the Radial Basis Function (RBF) and ploy kernel SVM.

Intuitively, the gamma parameter defines how far the influence of a single training example reaches, with low values meaning ‘far’ and high values meaning ‘close’. The gamma parameters can be seen as the inverse of the radius of influence of samples selected by the model as support vectors.

The C parameter trades off correct classification of training examples against maximization of the decision function’s margin. For larger values of C, a smaller margin will be accepted if the decision function is better at classifying all training points correctly. A lower C will encourage a larger margin, therefore a simpler decision function, at the cost of training accuracy. In other words``C`` behaves as a regularization parameter in the SVM. You can learn more about gamma and C at: https://scikit-learn.org/stable/auto_examples/svm/plot_rbf_parameters.html


```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler 

import matplotlib
import matplotlib.pyplot as plt
```

```python
def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True, which='both')
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)

def plot_predictions(clf, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)
```

```python
X, y = datasets.make_blobs(n_samples=100, centers=2)
plot_dataset(X, y, [-10, 20, -10, 15])
```

```python
from sklearn.svm import LinearSVC

lin_svc_pipeline = Pipeline([
    ('lin_svc', LinearSVC())
])

lin_svc_pipeline.fit(X, y)
plot_dataset(X, y, [-10, 20, -10, 15])
plot_predictions(lin_svc_pipeline, [-10, 20, -10, 15])
```

```python
from sklearn.datasets import make_moons
X_moons, y_moons = make_moons(n_samples=100, noise=0.15, random_state=42)
 
plot_dataset(X_moons, y_moons, [-1.5, 2.5, -1, 1.5])
```

#### Change the values of C and gamma below to observe behaviour of the classifier.

```python
from sklearn.svm import SVC


poly_kernel_svm_clf = Pipeline(( #Pipeline: Sequentially apply a list of transforms and a final estimator
            ("scaler", StandardScaler()), #Standardize features by removing the mean and scaling to unit variance
            ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=10, gamma=5))
        ))
poly_kernel_svm_clf.fit(X_moons, y_moons)
plot_predictions(poly_kernel_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X_moons, y_moons, [-1.5, 2.5, -1, 1.5])
```

```python
#Using a gaussian kernel (RBF kernel) to deal with non-linearity

rbf_kernel_svm_clf = Pipeline((
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="rbf", gamma=10, C=0.1))
    ))
rbf_kernel_svm_clf.fit(X_moons, y_moons)
plot_predictions(rbf_kernel_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X_moons, y_moons, [-1.5, 2.5, -1, 1.5])
```
