# Activation Functions

In this exercise we want to learn how some basic activation functions look like that were addressed in the lecture.
Try different values for a and b so that you understand their purpose.
What can you say about the derivatives of these functions?

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 10, 0.1) #create array of values

x
```

## Sigmoid function

```python
a = 1 #slope parameter

b = 0 #bias parameter

s = 1 / (1 + np.exp(-a * (x - b))) #sigmoid function

#plotting code
plt.plot(x, s)
plt.grid(b=True)
plt.show()
```

## Tanh function

```python
a = 1 #slope parameter

b = 0 #bias parameter

s = (2 / (1 + np.exp(-a * (x - b)))) - 1 #tanh function

#plotting code
plt.plot(x, s)
plt.grid(b=True)
plt.show()
```

## Heaviside function

```python
b = 0 #bias parameter

s = 1 * (x - b > 0) #step function

#plotting code
plt.step(x, s)
plt.grid(b=True)
plt.show()
```
